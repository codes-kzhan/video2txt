```python
import sys
import cv2
import glob
import os
        
```

```python
def process(file_name):
    vidcap = cv2.VideoCapture(file_name)
    success,image = vidcap.read()
    count = 0
    success = True
    while success:
        success,image = vidcap.read()
#        print 'Read a new frame: ', success
#        cv2.imwrite("frame%d.jpg" % count, image)     # save frame as JPEG file
        count += 1
    num_frames = count
    if(num_frames>10):
        num_frames /=10
    vidcap = cv2.VideoCapture(file_name)
    success,image = vidcap.read()
    success = True
    count = 0
    count1 = 0
    base_name =  os.path.basename(file_name)
    base_name = os.path.splitext(base_name)[0] 
    print("file_name -> %s" % base_name)
    while success:
        success,image = vidcap.read()
        if(count%num_frames==0):
            cv2.imwrite("./data/youtube/processed/%s_%d.jpg" % (base_name, count1), image)
            count1+=1
        count+=1
```

```python
dir_path = "./data/youtube/youtubeclips-dataset/"
for i in range(10,1971):
    file_name = dir_path+"vid"+str(i)+".avi"
    print (file_name)
    process(file_name)
```
