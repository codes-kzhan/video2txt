```python
import sys
import cv2
import glob
import os
import numpy as np
import pandas as pd
```

```python
#small_test = ['-4wsuPCjDBc','-7KMZQEsJW4','-8y1Q0rA3n8','-8y1Q0rA3n8','-9CUm-2cui8']
file_path = "./data/youtube/youtubeclips-dataset/video-descriptions.csv"
raw_data =  pd.read_csv(file_path)
file_path1 = "./data/youtube/youtubeclips-dataset/youtube_mapping.txt"
vid_map = [line.rstrip('\n') for line in open(file_path1)]
```

```python
def update(vid_id,vid_n1,vid_n2,vid_name):
    found = False
    for i in range(0,len(raw_data['id'])):
        if(str(raw_data['id'][i])==vid_id and str(raw_data['n1'][i])==vid_n1 and str(raw_data['n2'][i])==vid_n2):
            #print(raw_data['id'][i],raw_data['n1'][i],raw_data['n2'][i])
            raw_data.at[raw_data.index[i],'id'] = vid_name
            #print(raw_data['id'][i],raw_data['n1'][i],raw_data['n2'][i])
            found = True
            
    return found
```

```python
#vid_map = vid_map[0:5]
print len(vid_map)
#print vid_map
new_map = []
for line in vid_map:
    #print line
    parts = line.split(" ")
    vid_name = parts[1]
    parts1 = parts[0].split("_")
    vid_n1 = parts1[len(parts1)-2]
    vid_n2 = parts1[len(parts1)-1]
    vid_id = parts[0][0:11]
    
    #print vid_id, vid_n1, vid_n2
    get = update(vid_id,vid_n1,vid_n2,vid_name)
    if(get==False):
        print(line, vid_name)
    else:
        print(line,"Done")
```
