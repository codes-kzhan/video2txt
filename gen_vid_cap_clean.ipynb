```python
import sys
import cv2
import glob
import os
import numpy as np
import pandas as pd
```

```python
file_path = "./data/youtube/youtubeclips-dataset/small_vid_descriptions.csv"
raw_data =  pd.read_csv(file_path)
```

```python
def searchID(vid_id):
    cap = None
    print("-------------------------------------------------------------------------------")
    print(vid_id)
    for i in range(0,len(raw_data['id'])):
        if(str(raw_data['id'][i]) == vid_id and str(raw_data['status'][i]) == "clean"):
            cap = str(raw_data['caption'][i])
    return cap
```

```python
vocab = dict()
for i in range(1,1971):
    vid_id = "vid"+str(i)
    cap = searchID(vid_id)
    
    if(cap == None):
        print(vid_id)
    else:
        if(cap.endswith(".")):
            cap = cap[:-1]
        print cap
        
        vocab[vid_id] = cap
```

```python
for key,value in vocab.iteritems():
    vocab[key] = value.lower()
```
